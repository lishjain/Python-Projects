<h3>Analyzing Amazon order history</h3>
To download your order history CSV go to your Amazon account --> Accounts and Lists button in the top right --> On the next page, look for the Ordering and Shopping Preferences section, and click on the link under that heading that says “Download order reports”.

```python
import pandas as pd

df = pd.read_csv('amz_order_history.csv')
df.head();
```

```python
df.shape
```

```python
df = df.fillna(0)
df.head();
```

```python
df.dtypes
```

Changing the 'Total Charged' column to a float to proceed with calculations 

```python
df["Total Charged"] = df["Total Charged"].astype(float)
```

```python
df.dtypes
```

```python
df["Total Charged"].sum();
```

```python
df["Total Charged"].max();
```

```python
Max_value = df[df['Total Charged'] == df["Total Charged"].max()]
Max_value;
```

```python
df["Total Charged"].min();
```

```python
Min_value = df[df['Total Charged'] == df["Total Charged"].min()]
Min_value;
```

```python
df["Tax Charged"] = df["Tax Charged"].str.replace('$','').astype(float)
df.head();
```

```python
df["Tax Charged"].sum();
```
